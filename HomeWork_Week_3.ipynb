```python
# bieu do thanh nhom
# cho du lieu sau
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
labels = ["Q1", "Q2", "Q3", "Q4"]
A = [500,700,800,600]
B = [450,350,650,750]
C = [500,250,850, 600]
```

```python
# Ve bieu do thanh nhom the hien doanh thu cua A,B,C trong 4 quy
x = np.arange(len(labels))  # vị trí của các nhóm
width = 0.2  # Độ rộng của mỗi thanh

fig, ax = plt.subplots()
ax.bar(x - width, A, width, label='A')
ax.bar(x, B, width, label='B')
ax.bar(x + width, C, width, label='C')

ax.set_xlabel("Quý")
ax.set_ylabel("Doanh thu")
ax.set_title("Doanh thu của A, B, C trong 4 quý")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
# cho du lieu sau
# bieu do tron
Cty = ["A", "B", "C", "D"]
thiphan = [30,25,25,20]
color = ["#ff9999","#66b3ff","#99ff99","#ffcc99"]
```

```python
# tao bieu do tron bieu hien ti le thi phan cua 4 cong ty
from matplotlib import colors


plt.figure(figsize=(6, 6))
plt.pie(thiphan, labels=Cty, colors=color, autopct='%1.1f%%', startangle=140)
plt.title("Tỉ lệ thị phần của 4 công ty")
plt.show()
```

```python
# bieu do phan tan
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100)*300
colors = np.random.rand(100)
```

```python
# tao bieu do phan tan cua hai bien ngau nhien
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap='viridis')
plt.colorbar(label='Giá trị màu')
plt.xlabel("Biến X")
plt.ylabel("Biến Y")
plt.title("Biểu đồ phân tán của hai biến ngẫu nhiên")
plt.show()
```

```python
# Bieu do nhit
data = np.random.rand(10,10)
# ve mot heatmap the hien su phan bo cua du lieu tren ma tran
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap thể hiện sự phân bố của dữ liệu")
plt.show()
```
