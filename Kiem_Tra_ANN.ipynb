````python
import pandas as pd
import numpy as np

# Đọc dữ liệu
df = pd.read_csv('customer_data.csv')

# Bỏ các giá trị NaN trong cột Hours_Worked
hours_worked_clean = df['Hours_Worked'].dropna()

# Tính độ lệch chuẩn
std_hours = np.std(hours_worked_clean)
median_hours = np.median(hours_worked_clean)

print(f"Độ lệch chuẩn của Hours_Worked: {std_hours:.2f} giờ")
print(f"Trung vị (Median) của Hours_Worked: {median_hours} giờ")
````

Dề Xuất:
thiết lập khung giờ làm việc hợp lý dựa trên trung vị và kiểm soát giới hạn dao động dựa trên độ lệch chuẩn

Cụ thể:

Đặt giờ làm chuẩn mỗi tuần = median ± 1 std
Ví dụ: Nếu median = 41 và std ≈ 2.7, thì khung giờ chuẩn là từ 38.3 đến 43.7 giờ/tuần.

Nhân viên làm dưới 38h hoặc trên 44h nên được tư vấn/điều chỉnh lại khối lượng công việc.

Tăng giờ nghỉ giải lao, tự động cảnh báo nếu làm việc quá giới hạn trong tuần.

Thử nghiệm mô hình làm việc linh hoạt hoặc hybrid cho các vị trí phù hợp.

````python
# !pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv("customer_data.csv")

# Vẽ boxplot kết hợp Department và Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Performance', hue='Gender', data=df)
plt.title('Boxplot Performance theo Department và Gender')
plt.ylabel('Performance')
plt.xlabel('Department')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
# Loại bỏ dòng thiếu dữ liệu Performance
df_filtered = df.dropna(subset=['Performance'])

# Hàm tìm outliers theo IQR
def detect_outliers_IQR(sub_df):
    Q1 = sub_df['Performance'].quantile(0.25)
    Q3 = sub_df['Performance'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return sub_df[(sub_df['Performance'] < lower_bound) | (sub_df['Performance'] > upper_bound)]

# Tìm outliers theo từng nhóm Department + Gender
outliers = df_filtered.groupby(['Department', 'Gender'], group_keys=False).apply(detect_outliers_IQR)
print("Outliers được phát hiện:")
print(outliers[['Employee_ID', 'Department', 'Gender', 'Performance']])
````


Đề xuất
1. Chuẩn bị thông tin	
    - Tập hợp dữ liệu hiệu suất, đào tạo, giờ làm việc, và đánh giá của nhân viên outlier.
2. Gặp mặt cá nhân (1:1)	
    - Tạo không gian an toàn để thảo luận nguyên nhân (stress, thiếu kỹ năng, thiếu động lực, bất mãn, v.v.)
3. Hỏi & phân tích

4. Ghi nhận và đề xuất hỗ trợ	
    - Tư vấn thêm đào tạo, điều chỉnh khối lượng công việc, hoặc thay đổi nhóm nếu cần.
5. Theo dõi & đánh giá lại

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv('customer_data.csv')

# Điền dữ liệu thiếu
df['Performance'] = df['Performance'].fillna(df['Performance'].mean())
df['Training_Hours'] = df['Training_Hours'].fillna(df['Training_Hours'].mean())
df['Hours_Worked'] = df['Hours_Worked'].fillna(df['Hours_Worked'].mean())

# Tính trung bình Performance theo Department
department_performance = df.groupby('Department')['Performance'].mean().reset_index()

# Vẽ bar chart
plt.figure(figsize=(8, 5))
sns.barplot(data=department_performance, x='Department', y='Performance', palette='viridis', hue='Department', dodge=False)
plt.title('Average Performance by Department')
plt.xlabel('Department')
plt.ylabel('Average Performance')
plt.grid(axis='y')
plt.legend(title='Department', loc='upper right', frameon=False)
plt.show()
````

````python
from sklearn.linear_model import LinearRegression
import numpy as np

# Tạo mặt nạ cho các trường hợp có đầy đủ dữ liệu
du_lieu_mask = df[['Performance', 'Hours_Worked', 'Training_Hours']].notna().all(axis=1)
du_lieu_day_du = df[du_lieu_mask]

# Chuẩn bị đặc trưng (X) và mục tiêu (y) cho hồi quy
X = du_lieu_day_du[['Hours_Worked', 'Training_Hours']]
y = du_lieu_day_du['Performance']

# Xây dựng mô hình hồi quy tuyến tính
mo_hinh = LinearRegression()
mo_hinh.fit(X, y)

# In hệ số và hằng số
print("Các hệ số hồi quy:")
print(f"Số giờ làm việc: {mo_hinh.coef_[0]:.4f}")
print(f"Số giờ đào tạo: {mo_hinh.coef_[1]:.4f}")
print(f"Hằng số: {mo_hinh.intercept_:.4f}")

# Tạo bản sao của giá trị Performance
hieu_suat_du_doan = df['Performance'].copy()

# Với mỗi hàng có Performance bị thiếu
thieu_mask = df['Performance'].isna()
for idx in df[thieu_mask].index:
    # Lấy giá trị Hours_Worked và Training_Hours cho hàng này
    gio_lam = df.loc[idx, 'Hours_Worked']
    gio_dao_tao = df.loc[idx, 'Training_Hours']
    
    # Nếu có đủ cả hai biến dự đoán
    if not (np.isnan(gio_lam) or np.isnan(gio_dao_tao)):
        # Dự đoán Performance
        du_doan = mo_hinh.predict([[gio_lam, gio_dao_tao]])[0]
        hieu_suat_du_doan[idx] = du_doan

print("\nGiá trị Performance được dự đoán:")
print(hieu_suat_du_doan[thieu_mask])
````

````python
import pandas as pd

def identify_outliers_performance(df):
    # Tính trung bình và độ lệch chuẩn của Performance
    mean_performance = df['Performance'].mean()
    std_performance = df['Performance'].std()
    
    # Xác định ngưỡng
    lower_bound = mean_performance - 2 * std_performance
    upper_bound = mean_performance + 2 * std_performance
    
    # Tìm nhân viên có Performance ngoài 2 độ lệch chuẩn
    outliers = df[(df['Performance'] < lower_bound) | (df['Performance'] > upper_bound)]
    
    return outliers

# Đọc dữ liệu từ file CSV
df = pd.read_csv('customer_data.csv')

# Gọi hàm
outlier_employees = identify_outliers_performance(df)

# Hiển thị kết quả
print(outlier_employees)
````

Đề Xuất Chiến Lược Cho Batch Size

        - Tăng Độ Chính Xác và Khả Năng Tổng Quát: Kích thước batch nhỏ hơn có thể giúp mô hình cập nhật các trọng số thường xuyên hơn, dẫn đến việc tối ưu hóa tốt hơn và khả năng tổng quát cao hơn.
        - Giảm Thiểu Lượng Tài Nguyên: Với kích thước batch nhỏ, bạn có thể giảm lượng bộ nhớ cần thiết trong quá trình huấn luyện, cho phép bạn chạy trên các máy tính có tài nguyên hạn chế hơn.
    - Nhược Điểm:
        - Thời Gian Huấn Luyện Lâu Hơn: Sử dụng kích thước batch nhỏ có thể dẫn đến việc cần nhiều bước hơn để hoàn thành một epoch, do đó thời gian huấn luyện tổng thể có thể tăng lên.
        - Tính Không Ổn Định: Các cập nhật trọng số có thể trở nên không ổn định hơn với kích thước batch nhỏ, dẫn đến sự dao động lớn trong quá trình huấn luyện.
Kết Luận
- Việc chọn kích thước batch nhỏ hơn như 32 hoặc 16 có thể giúp cải thiện độ chính xác của mô hình trong khi vẫn giữ cho việc sử dụng tài nguyên hợp lý. Tuy nhiên, bạn cần cân nhắc giữa tốc độ huấn luyện và hiệu suất mô hình để tìm ra kích thước batch phù hợp nhất cho bài toán của mình.

## CÂU 18:

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt

# Định nghĩa các biến chuyển đổi
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Tải dữ liệu MNIST
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_subset, val_subset = random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(dataset=train_subset, batch_size=64, shuffle=True)
val_loader = DataLoader(dataset=val_subset, batch_size=64, shuffle=False)

# Định nghĩa kiến trúc ANN
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32, 10)

    def forward(self, x):
        x = x.view(-1, 28 * 28)  # Flatten
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x

model = ANN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
````

````python
# Hàm huấn luyện
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=20, patience=5):
    train_losses = []
    val_losses = []
    
    best_loss = float('inf')
    patience_counter = 0

    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        for images, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        
        avg_train_loss = total_loss / len(train_loader)
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for images, labels in val_loader:
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        avg_val_loss = val_loss / len(val_loader)
        val_losses.append(avg_val_loss)

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}')

        # Early stopping
        if avg_val_loss < best_loss:
            best_loss = avg_val_loss
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered")
                break

    return train_losses, val_losses

# Huấn luyện mô hình
train_losses, val_losses = train_model(model, train_loader, val_loader, criterion, optimizer)
````

````python
# Vẽ biểu đồ loss
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
````

Đề Xuất Kiến Trúc Khác
Kiến Trúc Đề Xuất
    - Số Lượng Nơ-ron: Tăng số nơ-ron trong các lớp ẩn lên 256, 128, và 64 nơ-ron.
    - Thêm Lớp Ẩn Khác: Thêm một lớp ẩn thứ tư với 32 nơ-ron.
Kiến Trúc Mới:

    - Input Layer: 784 nơ-ron
    - Hidden Layer 1: 256 nơ-ron
    - Hidden Layer 2: 128 nơ-ron
    - Hidden Layer 3: 64 nơ-ron
    - Hidden Layer 4: 32 nơ-ron
    - Output Layer: 10 nơ-ron
Giải Thích Lý Do
    - Tăng Cường Khả Năng Mô Hình: Việc tăng số nơ-ron trong các lớp ẩn có thể giúp mô hình học được nhiều đặc điểm phức tạp hơn từ dữ liệu.
    - Nâng Cao Khả Năng Tổng Quát: Thêm nhiều lớp và nơ-ron giúp mô hình có khả năng tổng quát tốt hơn cho các biến thể trong dữ liệu, đồng thời giúp giảm thiểu hiện tượng overfitting.
    - Tối Ưu Hóa Đào Tạo: Một kiến trúc sâu hơn có thể giúp cải thiện hiệu suất trong việc phân loại, đặc biệt với dữ liệu như MNIST, nơi có nhiều mẫu và sự biến đổi.
Kết Luận
- Bằng cách xây dựng mạng nơ-ron với kiến trúc được đề xuất và sử dụng early stopping, bạn có thể cải thiện khả năng học của mô hình. Việc điều chỉnh kiến trúc để phù hợp với dữ liệu cụ thể là rất quan trọng để đạt được hiệu suất cao hơn.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Hàm đánh giá mô hình
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_labels.extend(labels.numpy())
            all_preds.extend(predicted.numpy())

    accuracy = correct / total * 100
    return accuracy, all_labels, all_preds

# Đánh giá mô hình
accuracy, all_labels, all_preds = evaluate_model(model, val_loader)
print(f'Accuracy: {accuracy:.2f}%')

# Vẽ confusion matrix
cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(10))
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
````
